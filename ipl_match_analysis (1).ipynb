```python
df_mostruns = df_most_runs_and_strikerate[['batsman','total_runs']]
df_bestAvg = df_most_runs_and_strikerate[['batsman','average','numberofballs']]
df_beststrike = df_most_runs_and_strikerate[['batsman','average','strikerate','numberofballs']]

df_beststrike_rate = df_beststrike.groupby(['batsman','strikerate','average','numberofballs']).count().sort_values(by='strikerate',ascending=False)
df_best_strikerate_sorted = df_beststrike_rate.query('average > 35' and 'numberofballs > 3000').head(20)
df_best_strikerate_sorted
```

```python
df_most_runs = df_mostruns.groupby(['batsman','total_runs']).count().sort_values(by='total_runs',ascending=False)
df_most_runs
```

```python
df_best_Avg = df_bestAvg.groupby(['batsman','average','numberofballs']).count().sort_values(by='average',ascending=False)
sorted = df_best_Avg.query('numberofballs >= 3000')
sorted
```

```python
plt.figure(figsize=(4,4))
sns.lineplot(x= 'batsman' , y='average',data = sorted)
plt.xticks(rotation=90)
plt.title('Batting Average of Batsmen')
plt.show()


plt.figure(figsize=(4,4))
sns.lineplot(x='batsman' , y= 'strikerate' , data = df_best_strikerate_sorted)
plt.xticks(rotation=90)
plt.title('Batsman with best strike rate')
plt.show()
```
